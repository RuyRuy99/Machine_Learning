```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('stopwords') #baixar as stopwords
```

```python
count_vectorizer = CountVectorizer() #instancia o objeto
word_tokenizer = count_vectorizer.build_tokenizer()
word_tokenizer('eu gostei do filme. o filme é divertido')
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
stopwords
```

```python
count_vectorizer = CountVectorizer(stop_words=stopwords) #faz com que a vetorização retire as stopwords em pt-Br
word_tokenizer = count_vectorizer.build_analyzer() #quando vc tira as stopwords do CountVectorizer, vc instancia ele como build.analyzer
word_tokenizer('eu não gostei do filme. o filme é divertido')
```

# BoW

```python
dados = ["eu gostei do filme o filme é divertido",
         "eu não gostei do filme",
         "foi legal eu gostei",
         "filme do ano muito bom"]
```

```python
count_vectorizer = CountVectorizer()
```

```python
bag_of_words = count_vectorizer.fit_transform(dados) #tokeniza e transforma os dados de treino
```

```python
bag_of_words
```

```python
bag_of_words.toarray()
```

```python
count_vectorizer.get_feature_names() #retorna os tokens
```

```python
import pandas as pd
```

```python
pd.DataFrame(bag_of_words.toarray(),
             columns=count_vectorizer.get_feature_names())
```

```python
count_vectorizer.transform(['gostei filme']).toarray()
```

# TF-IDF

```python
# import tfidf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(dados)
```

```python
tfidf
```

```python
pd.DataFrame(tfidf.toarray(),
             columns=tfidf_vectorizer.get_feature_names())
```
