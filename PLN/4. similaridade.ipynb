```python
import pandas as pd
```

```python
df = pd.read_csv('B2W-Reviews01.csv')
```

```python
df['review_text']
```

```python
df = df.dropna().reset_index().drop(columns=['index'])
```

```python
reviews = df['review_text']
reviews
```

```python
#import do modulo
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#inicia a instancia numa variavel tfidf
tfidf = TfidfVectorizer()
```

```python
reviews_vec = tfidf.fit_transform(reviews) #tokeniza e transforma os dados
reviews_vec
#transforma numa matriz esparsa
```

```python
tfidf.get_feature_names()[5000:5010] #retorna os tokens
```

```python
#importar cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def newReview(review):
  review = tfidf.transform([review])
  #O que faz o transform aqui ?? Transforma os dados de teste
  cos = cosine_similarity(review, reviews_vec)#passa os dados de teste e o de treino
  
  return cos, np.argsort(cos)[0][::-1]#retorna a lista de tras pra frente
```

```python
cos, idx = newReview('brinquedo')
```
