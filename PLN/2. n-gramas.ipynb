```python
bigrams = []

for i in range(len(words)-1):
    bigrams.append(words[i] + ' ' + words[i+1])# contatenando duas strings e adicionando na lista.

bigrams[0:10]
```

```python
def p_bigram(w1, w2, unigrams, bigrams):
    unigrams_freq = Counter(unigrams) 
    bigrams_freq = Counter(bigrams)
    return bigrams_freq[w1 + ' ' + w2] / unigrams_freq[w1] * 100
```

```python
p_bigram('ajudou', 'a', words, bigrams)
```

```python
p_bigram('e', 'capitu', words, bigrams)
```

```python
p_bigram('por', 'favor', words, bigrams)
```

```python
def p_trigrama(w1,w2,w3,unigrama,bigrama,trigrama):
  unigrama_freq = Counter(unigrama)
  bigrama_freq = Counter(bigrama)
  trigrama_freq = Counter(trigrama)
  #o return está certo ??
  #Não se usa unigrama na formulação do trigrama completo ?
  return trigrama_freq[w1 + " " + w2 + " " + w3] / bigrama_freq[w1 + " " + w2]

print(p_trigrama("de","si","mesmos",words,bigrams,trigramas))
```

```python
def shannon(bigrams,n_words):
  frase = []

  first = random.choice(bigrams).split(" ")#sleeciona um bigrama aleatoriamente e da split nas palavras
  #first = str(input("Digite um bigrama :   ")).split(" ")
  w1 = frase.append(first[0])
  w2 = frase.append(first[1])

  for i in range(n_words): #vai repetir n palavras
    filtered_bigrams = [] #lista com os bigramas que começam com a última palavra da lista "frase"
    for b in bigrams: #varre a lista de bigramas
      palavras = b.split(" ")
      w1 = palavras[0]
      w2 = palavras[1]

      if(w1 == frase[-1]):#se a primeira palavra do prox. bigrama for igual a ultima palavra da lista "frase"
        filtered_bigrams.append(b)

    novo_bigrama = random.choice(filtered_bigrams).split(" ") #pega um bigrama aleatorio da lista de bigramas
    frase.append(novo_bigrama[1]) #adiciona apenas a ultima palavra a lista final

  return " ".join(frase)
```

```python
import time
inicio = time.time()
print(shannon(bigrams,20))
fim = time.time()
print(fim - inicio)
```
