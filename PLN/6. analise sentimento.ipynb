```python
import pandas as pd
```

```python
df = pd.read_csv('../../datasets/B2W-Reviews01.csv')
df
```

```python
df['overall_rating'].value_counts()
```

```python
%time df['label'] = df['overall_rating'].apply(lambda x: 'pos' if x >= 3 else 'neg')
```

```python
import numpy as np

%time df['label'] = np.where(df['overall_rating'] >= 3, 'pos', 'neg')
```

```python
df[['overall_rating', 'label']]
```

```python
df['label'].value_counts()
```

```python
df.dropna(inplace=True, subset=['review_text'])
```

```python
df
```

```python
# import countvectorizer
from sklearn.feature_extraction.text import CountVectorizer

# instantiate countvectorizer
vect = CountVectorizer()

# fit and transform
review_vect = vect.fit_transform(df['review_text'])
```

```python
# import train test split
from sklearn.model_selection import train_test_split

# split data
X_train, X_test, y_train, y_test = train_test_split(review_vect, df['label'], test_size=0.2)
```

```python
# import naive bayes
from sklearn.naive_bayes import MultinomialNB

# instantiate naive bayes   
nb = MultinomialNB()

# fit model
nb.fit(X_train, y_train)
```

```python
nb.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, nb.predict(X_test)))
```

```python
my_input = ['muito']

my_input_vect = vect.transform(my_input)

nb.predict(my_input_vect)
```

```python
# show naive bayes probabilities for each word
feature_probs = nb.feature_log_prob_
feature_probs
```
